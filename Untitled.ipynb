```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
base = pd.read_csv('Eleicao.csv', sep=';')
plt.scatter(base.DESPESAS, base.SITUACAO)
base.describe()
```

```python
np.corrcoef(base.DESPESAS, base.SITUACAO)
```

```python
y = base.iloc[:,1].values  # Situacao
x = base.iloc[:,2].values  # Despesas
x = x[:,np.newaxis] # x virá matriz e add um novo eixo
x
```

```python
# Criação do modelo de treinamento
modelo = LogisticRegression()
modelo.fit(x,y)
modelo.coef_ # inclinação
modelo.intercept_  # interceptação
```

```python
plt.scatter(x,y)
# Gerando novos dados para função sigmoide
x_teste = np.linspace(10, 3000, 100)
# Implementação da função sigmoide
def model(x):
  return 1 / (1 + np.exp(-x))
# Geração de previsões (variável R) e visualização dos resultados
r = model(x_teste + modelo.coef_ + modelo.intercept_).ravel()
plt.plot(x_teste, r, color = 'purple')
```
